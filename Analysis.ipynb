```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("data/Extended_Employee_Performance_and_Productivity_Data.csv")
```

```python
data.head()
```

```python
data.drop(columns=['Employee_ID', 'Hire_Date', 'Work_Hours_Per_Week', 'Sick_Days', 'Remote_Work_Frequency', 'Team_Size', 'Training_Hours', 'Resigned'], inplace=True)
```

```python
data
```

```python
data.isna().sum().sum()
```

```python
data.duplicated().sum()
```

```python
data.groupby("Department")["Monthly_Salary"].mean().sort_values(ascending=False).plot(kind="bar",colormap="plasma")
plt.title("Average Salaries by Department")
plt.ylabel("Average Salary")
plt.xlabel("Department")
plt.show()
```

```python
data["Performance_Score"].value_counts().plot(kind="pie", colors=["#FF9999", "#66B3FF", "#99FF99", "#FFB366"])
plt.title("Performance Score Counts")
plt.ylabel("")
plt.legend()
plt.show()

```
